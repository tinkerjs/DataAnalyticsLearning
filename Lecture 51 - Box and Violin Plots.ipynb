```python
#The normal imports
import pandas as pd
import numpy as np
from numpy.random import randn

#Import the stats library from numpy
from scipy import stats

#These are the plotting modules and libs we will use
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

#Command so that plots appear in the iPython Notebook
%matplotlib inline
```

```python
data1 = randn(100)
data2 = randn(100)
```

```python
sns.boxplot([data1,data2])
```

```python
sns.boxplot([data1,data2], whis=np.inf)
```

```python
sns.boxplot([data1,data2],vert=False)
```

```python
#Normal Dist
data3 = stats.norm(0,5).rvs(100)

# Two gamma distributions. Contatonated together
data4 = np.concatenate([stats.gamma(5).rvs(50)-1, 
                       -1*stats.gamma(5).rvs(50)])

#boxplot data 1 and data2

sns.boxplot(data=[data3,data4])
```

```python
sns.violinplot(data=[data3,data4])
```

```python
sns.violinplot(data=[data3],bw=0.01)
```
