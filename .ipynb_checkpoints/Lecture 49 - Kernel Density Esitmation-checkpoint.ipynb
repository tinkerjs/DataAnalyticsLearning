```python
#The normal imports
import pandas as pd
import numpy as np
from numpy.random import randn

#Import the stats library from numpy
from scipy import stats

#These are the plotting modules and libs we will use
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

#Command so that plots appear in the iPython Notebook
%matplotlib inline
```

```python
#rug plot. Puts a tick wherever a value occurs

dataset = randn(25)
sns.rugplot(dataset)

plt.ylim(0,1)
```

```python
plt.hist(dataset, alpha=.3)
sns.rugplot(dataset)
```

```python
sns.rugplot(dataset)

x_min = dataset.min() - 2
x_max = dataset.max() + 2

x_axis = np.linspace(x_min, x_max, 100)

bandwidth = ((4*dataset.std()**5)/(3*len(dataset))) **0.2

kernel_list = []

for data_point in dataset:
    #create a kernel for each point and append it to the kernel_list
    kernel = stats.norm(data_point, bandwidth).pdf(x_axis)
    kernel_list.append(kernel)
    
    #scale for plotting
    kernel = kernel / kernel.max()
    kernel = kernel * .4
    
    plt.plot(x_axis, kernel, color='grey', alpha=0.5)
    
plt.ylim(0,1)
```

```python
sum_of_kde = np.sum(kernel_list, axis=0)

fig = plt.plot(x_axis, sum_of_kde, color='indianred')

sns.rugplot(dataset)

plt.yticks([])

plt.suptitle('Sum of the basis functions')
```

```python
#using seaborn
sns.kdeplot(dataset)
```

```python
sns.rugplot(dataset, color='black')

for bw in np.arange(0.5,2,.25):
    sns.kdeplot(dataset, bw=bw, lw=1.8, label=bw)
```

```python
kernel_options = ['biw','cos','epa','gau','tri','triw']

for kern in kernel_options:
    sns.kdeplot(dataset, kernel=kern, label=kern, shade = True)
```

```python
sns.kdeplot(dataset, vertical=True)
```

```python
#cumulative distribution function

sns.kdeplot(dataset, cumulative = True)
```

```python
mean = [0,0]

cov = [[1,0],[0,100]]

dataset2 = np.random.multivariate_normal(mean,cov,1000)

dframe = pd.DataFrame(dataset2, columns=['X','Y'])

sns.kdeplot(dframe)
```

```python
sns.kdeplot(dframe.X, dframe.Y, shade=True)
```

```python
sns.kdeplot(dframe,bw=1)
```

```python
sns.kdeplot(dframe,bw='silverman')
```

```python
sns.jointplot('X','Y', dframe, kind='kde')
```
